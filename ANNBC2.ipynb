```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.metrics import accuracy_score, mean_squared_error


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

from livelossplot import PlotLossesKeras
```

```python
#Load the first dataset
dataset = pd.read_csv('data_refined.csv')
dataset
```

```python
#I built this before learning how to not have it add a column so i'll drop that column back out

dataset = dataset.drop(["Unnamed: 0"], axis = 1)
dataset
```

```python
# and then train/test/split
encoded_y = dataset.iloc[:, 0].values
encoded_x = dataset.iloc[:, 1:21].values

# train is now 80% of the entire data set
X_train, X_test, y_train, y_test = train_test_split(encoded_x, encoded_y, test_size=0.2)

print(len(X_train), len(X_test))
# test is now 10% of the initial data set
# validation is now 10% of the initial data set
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=.5) 
print(len(X_train), len(X_val), len(X_test))
```

Typically we would scale AFTER the train test split, however as we are building from a previous project where we hadn't gotten to that part yet in this instance for the time being I will use the existing data as is.  Possibly can go back and redo and then do scaling at this point.

```python
# Next to set up the ANN MLPClassifier - we were not advised how many hidden layers to choose, so using 42 as per the notebook example for a simple ANN as a start point, and changing max iterations to 1000
mlp_classifier = MLPClassifier(hidden_layer_sizes=(42), 
                               max_iter=1000, 
                               random_state=0)

mlp_classifier.fit(X_train, y_train)

```

```python
mlp_pred = mlp_classifier.predict(X_test)
mlp_pred
```

```python
mlp_accuracy = accuracy_score(y_test, mlp_pred)
print("Model Accuracy:", mlp_accuracy)

# We have successfully with these parameters gotten the hoped for minimum of 94%
```

```python
# Next we will do Keras

# Set-up the architecture for a TensorFlow MLP Classifier
tf_classifier = Sequential()

tf_classifier.add(Dense(50, activation='relu'))
tf_classifier.add(Dense(4, activation='softmax'))
```

```python
#Compile the network with chosen hyperparameters - note that no specific paramaters were given in assignment so I may try one with the loss type used in the notebook and another with loss type binary_crossentropy
tf_classifier.compile(loss='sparse_categorical_crossentropy', 
                      optimizer='adam', 
                      metrics=['accuracy'])
```

```python
# Add early stopping
callbacks = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                              min_delta=0.0001,
                              patience=5,
                              verbose=1, mode='auto')
```

```python
#Train the model
tf_classifier.fit(X_train,
                  y_train, 
                  epochs=1000, validation_data=(X_val, y_val), 
                  callbacks=[PlotLossesKeras(),callbacks],
                  verbose=1)  # removed PlotLossesKeras(), from callbacks for now due to error will look further into.
```

```python
#Make predictions on the test data with the trained model
tf_class_pred = np.argmax(tf_classifier.predict(X_test), axis=1)
tf_class_pred
```

```python
tf_class_accuracy = accuracy_score(y_test, tf_class_pred)
print("Model Accuracy:", tf_class_accuracy)
```
