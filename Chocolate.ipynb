```python
#Convert the Column Percent.
#Change the type of values in the column percent from string values to numerical values.  
```

```python
dataset['CocoaPercent'] = dataset['CocoaPercent'].str.rstrip('%').astype('float')
dataset
```

```python
# Normalize the Rating column

# copy the data 
dataset_max_scaled = dataset.copy() 
  
# apply normalization techniques on Rating
column = 'Rating'
dataset_max_scaled[column] = dataset_max_scaled[column] /dataset_max_scaled[column].abs().max() 
  
# view normalized data 
display(dataset_max_scaled) 
```

```python
x=dataset['Rating']
y=dataset['CocoaPercent']
plt.scatter(x, y)
plt.title('Ratings Vs Cocoa Percentage')
plt.xlabel("Rating")
plt.ylabel("Percentage of Cocoa")
plt.show()
```
